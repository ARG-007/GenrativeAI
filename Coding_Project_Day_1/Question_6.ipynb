### Write a program that uses cv2 and rotates an image. The code will create three images: "tester1.png", "tester2.png", and "tester3.png". Each image will be a variation of the original image based on the applied pixel assignment patterns such that the resulting image is rotated at 90°, 180°, 270° respectively. Upload screenshots of both the code and the output (upload only in PNG image format)

```python
# Preprocessing
import cv2

originalImage = cv2.imread(r"./program_io/Skyrim_Refashioned_Logo_1.png")
rgbOImg = cv2.cvtColor(originalImage, cv2.COLOR_BGR2RGB)
print(originalImage.shape)

height, width, colors = originalImage.shape

imageCenter = ((width - 1) / 2.0, (height - 1) / 2.0)

rotate_mat_90 = cv2.getRotationMatrix2D(imageCenter, 90.0, 1)
```

```python
image_rotated_90 = cv2.warpAffine(originalImage, rotate_mat_90, (width, height))
image_rotated_180 = cv2.warpAffine(image_rotated_90, rotate_mat_90, (width, height))
image_rotated_270 = cv2.warpAffine(image_rotated_180, rotate_mat_90, (width, height))
```

```python
import matplotlib.pyplot as plot

fig, axs = plot.subplots(2, 2, squeeze=True, figsize=(10, 10), dpi=400)

axs[0, 0].imshow(cv2.cvtColor(originalImage, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title("Original Image")

axs[0, 1].imshow(cv2.cvtColor(image_rotated_90, cv2.COLOR_BGR2RGB))
axs[0, 1].set_title("Image Rotated 90")

axs[1, 0].imshow(cv2.cvtColor(image_rotated_180, cv2.COLOR_BGR2RGB))
axs[1, 0].set_title("Image Rotated 180")

axs[1, 1].imshow(cv2.cvtColor(image_rotated_270, cv2.COLOR_BGR2RGB))
axs[1, 1].set_title("Image Rotated 270")

for axx in axs:
    for axy in axx:
        axy.set_xticks([])
        axy.set_yticks([])

plot.subplots_adjust(hspace=-0.5)
plot.savefig(fname=r"./program_io/q6_output.png", format="png", dpi=500)
plot.show()
```

```python
cv2.imwrite(r"./program_io/img90.png", image_rotated_90)
cv2.imwrite(r"./program_io/img180.png", image_rotated_180)
cv2.imwrite(r"./program_io/img270.png", image_rotated_270)
```
